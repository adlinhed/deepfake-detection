NOTEBOOK: Split dataset, create folders, move images

```python
import os
import shutil
from sklearn.model_selection import train_test_split
from PIL import Image

base_dir = '/kaggle/input/datasets/ciplab/real-and-fake-face-detection/real_and_fake_face' 
output_dir = "/kaggle/working/dataset_split"  
```

```python
classes = os.listdir(base_dir)
print("Detected classes:", classes)

for cls in classes:
    os.makedirs(os.path.join(output_dir, "train", cls), exist_ok=True)
    os.makedirs(os.path.join(output_dir, "val", cls), exist_ok=True)
    os.makedirs(os.path.join(output_dir, "test", cls), exist_ok=True)

# Split + copy
for cls in classes:
    class_path = os.path.join(base_dir, cls)
    images = os.listdir(class_path)

    train_imgs, temp_imgs = train_test_split(images, test_size=0.30, random_state=42)
    val_imgs, test_imgs = train_test_split(temp_imgs, test_size=0.50, random_state=42)

    def copy_valid_images(img_list, split_name):
        for img in img_list:
            src = os.path.join(class_path, img)
            dst = os.path.join(output_dir, split_name, cls, img)

            if not img.lower().endswith(('.png', '.jpg', '.jpeg')):
                continue

            try:
                Image.open(src).verify()
                shutil.copy(src, dst)
            except:
                continue

    copy_valid_images(train_imgs, "train")
    copy_valid_images(val_imgs, "val")
    copy_valid_images(test_imgs, "test")

print("Dataset split complete.")
```
